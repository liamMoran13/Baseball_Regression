```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('baseball_players.xls')
```

```python
df = df[df['Salary'].notna()]
df['Salary'] = df['Salary'].str.replace('$','').str.replace(',','').astype(int)
```

```python
first_viz = sns.histplot(df['Salary'])
plt.title('Histogram of Salaries',fontsize=20)
sns.set_style('darkgrid')
plt.savefig('Histogram_of_Salaries.png',bbox_inches='tight')
```

```python
first_viz = sns.histplot(np.log(df['Salary']))
plt.title('Histogram of Salaries Log Transformed',fontsize=20)
sns.set_style('darkgrid')
```
